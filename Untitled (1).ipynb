```python
class Employee:
    def __init__(self, name, number, address, salary, post, year_of_joining, education, year_of_experience):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.year_of_experience = year_of_experience

    def describe_employee(self):
        print(f"Name: {self.name}")
        print(f"Number: {self.number}")
        print(f"Address: {self.address}")
        print(f"Salary: {self.salary}")
        print(f"Post: {self.post}")
        print(f"Year of Joining: {self.year_of_joining}")
        print(f"Education: {self.education}")
        print(f"Years of Experience: {self.year_of_experience}")

    def update_year_of_experience(self):
        self.year_of_experience += 1

    def increase_salary(self, amount):
        self.salary += amount


# Create an instance of Employee
employee101 = Employee("John Doe", "E123", "123 Elm Street", 50000, "Software Engineer", 2020, "BSc Computer Science", 4)

# Print attributes individually
print(employee101.name)
print(employee101.number)
print(employee101.address)
print(employee101.salary)

# Call methods
employee101.describe_employee()
employee101.increase_salary(5000)
employee101.update_year_of_experience()

# Print updated attributes
print("\nUpdated Information:")
employee101.describe_employee()

# Create 10 different instances
employees = [
    Employee("Alice Smith", "E124", "456 Oak Street", 55000, "Data Analyst", 2019, "MSc Statistics", 5),
    Employee("Bob Johnson", "E125", "789 Pine Street", 60000, "Product Manager", 2018, "MBA", 6),
    Employee("Carol Davis", "E126", "321 Maple Street", 65000, "HR Specialist", 2021, "BBA", 3),
    Employee("Dave Wilson", "E127", "654 Cedar Street", 70000, "UX Designer", 2017, "MA Design", 7),
    Employee("Eve Martinez", "E128", "987 Birch Street", 75000, "Finance Manager", 2015, "CPA", 9),
    Employee("Frank Brown", "E129", "111 Elm Street", 80000, "Software Architect", 2016, "PhD Computer Science", 8),
    Employee("Grace Taylor", "E130", "222 Oak Street", 85000, "Marketing Director", 2014, "MBA Marketing", 10),
    Employee("Hank Anderson", "E131", "333 Pine Street", 90000, "Sales Manager", 2018, "BCom", 6),
    Employee("Ivy Thomas", "E132", "444 Maple Street", 95000, "Customer Support Lead", 2020, "BSc Psychology", 4),
    Employee("Jack Lee", "E133", "555 Cedar Street", 100000, "Business Analyst", 2017, "MSc Business Analytics", 7)
]

# Call describe_employee() for each instance
for emp in employees:
    emp.describe_employee()
    print()

# Perform some updates with its methods
for emp in employees:
    emp.increase_salary(3000)
    emp.update_year_of_experience()
    
# Display updated information for each employee
for emp in employees:
    emp.describe_employee()
    print()
```
